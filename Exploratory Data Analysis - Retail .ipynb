```python
df_state=df.groupby(['State'])[['Quantity','Sales','Discount','Profit']].mean()
df_state.head()
```

```python
df_state=df.groupby(['State'])[['Quantity','Sales','Discount','Profit']].describe()
df_state.head()
```

```python
sp=df.groupby('State')['Profit'].sum()
sp.head()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x=sp.index, y=sp.values)
plt.xticks(rotation=60)
plt.ylabel('Profit')
plt.show()
```

```python
sns.pairplot(df, hue='Region')
```

```python
sns.pairplot(df, hue='Category')
```

```python
sns.pairplot(df,hue='Category',diag_kind='hist')
```

```python
sns.countplot('Segment', data=df)
```

```python
state = df['State'].value_counts()
plt.figure(figsize=(15,16))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,10))
df.groupby('Category')['Profit'].sum().plot.pie(autopct='%1.1f%%', shadow=True)
plt.title('Category Wise Profit Analysis', fontsize=20)
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Category', hue='Region', data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
order=df.groupby(['State', 'City']).size().reset_index(name='total_orders')
order[order['total_orders']==order['total_orders'].min()]
```

```python
plt.figure(figsize=(15,13))
df.groupby('State')['Sales'].sum().plot.pie(autopct='%1.1f%%')
plt.title('Sale Analysis', fontsize=20)
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize = (15,13))
sns.distplot(df['Sales'], ax = axs[0,0])
sns.distplot(df['Quantity'], ax = axs[0,1])
sns.distplot(df['Discount'], ax = axs[1,0])
sns.distplot(df['Profit'], ax = axs[1,1])
```

```python
pd.crosstab(df['Region'], df['Category']).plot(kind='bar', figsize=(15,7))
plt.xticks(rotation=90)
plt.plot()
```
